```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

# Set font settings
matplotlib.rcParams["mathtext.fontset"] = "cm"
matplotlib.rcParams["font.family"] = "STIXGeneral"

#  Load each dataset separately 
nxx = np.loadtxt('data/nxx.csv', delimiter='\t')
nH = np.loadtxt('data/Higgs.csv', delimiter='\t')
nyy = np.loadtxt('data/nyy.csv', delimiter='\t')
nbare=nxx-nH;
# Load the x-axis values
xaxis = np.loadtxt('data/xvalue.csv', delimiter='\t')

# Create subplots (2 rows, 1 column)
fig, axs = plt.subplots(2, 1, figsize=(8, 10))

# Plot the nxx dataset in the first subplot
axs[0].plot(xaxis, nxx/0.283812, '-',linewidth=5, label=r'$n_s^{xx}(T)$',color="dodgerblue")
axs[0].plot(xaxis, nyy/0.283812, '-',linewidth=5, color="orange",label=r"$n_s^{yy}(T)$")

#axs[0].set_title(r'Superfluid Density',fontsize=20)
 
axs[0].plot(xaxis, xaxis*0,  linestyle="dotted",linewidth=3,color="darkred")


axs[0].set_ylim([-0.06,0.1])

axs[0].tick_params(axis='x', labelsize=20)  # Change font size of x-axis labels
axs[0].tick_params(axis='y', labelsize=20)  # Change font size of x-axis labels

axs[0].legend(fontsize=20)
axs[0].text(0.01/2,0.9,r"(a)",fontsize=20)
# Plot the nyy dataset in the second subplot


#axs[0].plot(xaxis[10], nyy[10]/0.2889, 'o',markersize=10, color="orange")


#axs[1].set_title('Plot of nyy vs x')
#axs[0].set_xlabel(r'$T/t$',fontsize=25)
axs[0].set_ylabel(r'$n_s(T)/n_D$',fontsize=25)
axs[0].legend(fontsize=20)
axs[0].set_xlim([0,0.2])
axs[0].set_xticks([0, 0.1,0.2])
axs[0].set_yticks([0,0.5,1])
axs[0].tick_params(axis='x', labelsize=20)  # Change font size of x-axis labels
axs[0].tick_params(axis='y', labelsize=20)  # Change font size of x-axis labels


axs[0].annotate(
    "",
    xy=(0.05, 0.75),
    xytext=(0.05, 0.95),
    arrowprops=dict(
        arrowstyle="-|>",
        linewidth=5,
        edgecolor="orange",
        facecolor="orange",
        mutation_scale=40
    )
)

axs[0].annotate(
    "",
    xy=(0.1, 0.15),
    xytext=(0.1, 0.95),
    arrowprops=dict(
        arrowstyle="-|>",
        linewidth=5,
        edgecolor="cornflowerblue",
        facecolor="cornflowerblue",
        mutation_scale=40
    )
)

### T^2
axs[1].plot(xaxis, nH/0.283812, label=r'$n^{xx}_{H}$',linewidth=5,linestyle="dotted",color="dodgerblue")

axs[1].plot(xaxis, nbare/0.283812, label=r'$n^{xx}_0$',linewidth=5,linestyle="dashed",color="dodgerblue")

axs[1].text(0.01/2,0.1-0.01,r"(b)",fontsize=20)

axs[1].set_xlabel(r'$T/t$',fontsize=25)
axs[1].set_ylabel(r'$n_s^{xx}/n_D$',fontsize=25)
axs[1].set_xlim([0,0.2])
#axs[0].set_xticks([0, 0.2, 0.4])
axs[1].set_yticks([-0.1,0,0.1])
axs[1].set_xticks([0, 0.1,0.2])
#axs[1].set_ylim([-0.06,0.02])

axs[1].tick_params(axis='x', labelsize=20)  # Change font size of x-axis labels
axs[1].tick_params(axis='y', labelsize=20)  # Change font size of x-axis labels

axs[1].legend(fontsize=20)

 
plt.savefig("fig2.pdf", bbox_inches='tight')
 
```
